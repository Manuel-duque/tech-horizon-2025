```python
cuadro3 = df["country"].value_counts().reset_index()
cuadro3.columns = ["País", "Cantidad de circuitos"]

print(" Cuadro 3: Cantidad de circuitos por país")
print(cuadro3.head(10))  # Muestra los 10 primeros
```

```python
def clasificar_altitud(x):
    if pd.isna(x):
        return "Desconocida"
    elif x < 100:
        return "Baja"
    elif x <= 500:
        return "Media"
    else:
        return "Alta"


df["categoria_altitud"] = df["alt"].apply(clasificar_altitud)

print(" Nueva columna creada: 'categoria_altitud'")
print(df[["name", "country", "alt", "categoria_altitud"]].head(10))
```
