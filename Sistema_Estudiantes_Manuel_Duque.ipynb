```python
# === Sistema de Estudiantes Mejorado ===
# Lista donde se guardan todos los estudiantes
estudiantes = []

# Función para agregar un nuevo estudiante
def agregar_estudiante(nombre, edad, grado, materias, promedio):
    # Se crea un diccionario con los datos del estudiante
    estudiante = {
        "nombre": nombre,
        "edad": edad,
        "grado": grado,
        "materias": materias,
        "promedio": promedio
    }
    # Se agrega el estudiante a la lista principal
    estudiantes.append(estudiante)
    print(f"Estudiante '{nombre}' agregado correctamente.\n")

# Función para eliminar un estudiante por su nombre
def eliminar_estudiante(nombre):
    for estudiante in estudiantes:
        # Se compara el nombre sin importar mayúsculas o minúsculas
        if estudiante["nombre"].lower() == nombre.lower():
            estudiantes.remove(estudiante)
            print(f"Estudiante '{nombre}' eliminado.\n")
            return
    # Si no se encuentra el estudiante, se muestra un mensaje
    print("Estudiante no encontrado.\n")

# Función para actualizar la información de un estudiante
def actualizar_estudiante(nombre):
    estudiante = buscar_estudiante(nombre)
    if estudiante:
        print("=== Actualizar información ===")
        # Se piden los nuevos datos del estudiante
        estudiante["edad"] = int(input("Nueva edad: "))
        estudiante["grado"] = input("Nuevo grado: ")
        estudiante["materias"] = input("Nuevas materias favoritas (separadas por comas): ").split(",")
        estudiante["promedio"] = float(input("Nuevo promedio: "))
        print("Información actualizada con éxito.\n")
    else:
        print("Estudiante no encontrado.\n")

# Función para buscar un estudiante por nombre
def buscar_estudiante(nombre):
    for estudiante in estudiantes:
        if estudiante["nombre"].lower() == nombre.lower():
            return estudiante
    return None

# Función para calcular el promedio general de todos los estudiantes
def calcular_promedio_general():
    if not estudiantes:
        return 0  # Si la lista está vacía, devuelve 0
    # Suma todos los promedios y divide entre la cantidad de estudiantes
    total = sum(est["promedio"] for est in estudiantes)
    return total / len(estudiantes)

# Función para mostrar todos los estudiantes registrados
def mostrar_todos():
    if not estudiantes:
        print("No hay estudiantes registrados.\n")
        return
    print("\n=== LISTA DE ESTUDIANTES ===")
    for est in estudiantes:
        print(f"{est['nombre']} | Edad: {est['edad']} | Grado: {est['grado']} | "
              f"Materias: {', '.join(est['materias'])} | Promedio: {est['promedio']}")
    print()

# === Ejemplos de estudiantes ya agregados ===
# Se agregan automáticamente al iniciar el programa
agregar_estudiante("Ana", 15, "10°", ["Matemáticas", "Ciencias"], 4.5)
agregar_estudiante("Luis", 16, "11°", ["Historia", "Inglés"], 3.8)
agregar_estudiante("María", 15, "10°", ["Arte", "Educación Física"], 4.2)
agregar_estudiante("Carlos", 17, "12°", ["Física", "Química"], 4.7)
agregar_estudiante("Sofía", 16, "11°", ["Música", "Matemáticas"], 3.9)

# === Menú interactivo ===
def menu():
    while True:  # El menú se repite hasta que el usuario elija salir
        print("=== MENÚ PRINCIPAL ===")
        print("1. Agregar estudiante")
        print("2. Eliminar estudiante")
        print("3. Actualizar información")
        print("4. Mostrar todos los estudiantes")
        print("5. Calcular promedio general")
        print("6. Salir")

        # El usuario elige una opción
        opcion = input("Elige una opción: ")

        # Opción 1: agregar estudiante nuevo
        if opcion == "1":
            nombre = input("Nombre: ")
            edad = int(input("Edad: "))
            grado = input("Grado: ")
            materias = input("Materias favoritas (separadas por comas): ").split(",")
            promedio = float(input("Promedio: "))
            agregar_estudiante(nombre, edad, grado, materias, promedio)

        # Opción 2: eliminar estudiante
        elif opcion == "2":
            nombre = input("Nombre del estudiante a eliminar: ")
            eliminar_estudiante(nombre)

        # Opción 3: actualizar datos
        elif opcion == "3":
            nombre = input("Nombre del estudiante a actualizar: ")
            actualizar_estudiante(nombre)

        # Opción 4: mostrar todos los estudiantes
        elif opcion == "4":
            mostrar_todos()

        # Opción 5: calcular promedio general del grupo
        elif opcion == "5":
            promedio_general = calcular_promedio_general()
            print(f"Promedio general del grupo: {promedio_general:.2f}\n")

        # Opción 6: salir del programa
        elif opcion == "6":
            print("Saliendo del programa...")
            break

        # Si la opción no existe, muestra un mensaje
        else:
            print("Opción inválida. Intenta de nuevo.\n")

# Ejecutar el menú principal
menu()
```
